```python
df1 = pd.read_csv('2018-04-25_11-38-56.csv', index_col=0)
df2 = pd.read_csv('2018-04-25_11-39-17.csv', index_col=0)
df4 = pd.read_csv('2018-04-25_11-39-33.csv', index_col=0)
df8 = pd.read_csv('2018-04-25_11-39-46.csv', index_col=0)
df16 = pd.read_csv('2018-04-25_11-40-07.csv', index_col=0)
df32 = pd.read_csv('2018-04-25_11-40-29.csv', index_col=0)
```

```python
toma = pd.concat([df1.iloc[:,0],
                  df2.iloc[:,0],
                  df4.iloc[:,0],
                  df8.iloc[:,0],
                  df16.iloc[:,0],
                  df32.iloc[:,0]], axis=1)
toma.columns = ['Nodes 1', 'Nodes 2', 'Nodes 4', 'Nodes 8', 'Nodes 16', 'Nodes 32']
toma.head()
```

# Convergence

```python
fig, ax = plt.subplots(nrows=1, ncols=1, sharex=True)
toma.iloc[::20].plot(ax=ax, logy=True, ms=5, style=['.', '.', '.', '.', '.', '.'])
ax.set_ylabel('Relative error')

# fig.savefig('toma.png', dpi=300)
```
